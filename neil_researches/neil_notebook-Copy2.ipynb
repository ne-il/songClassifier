```python
import numpy as np
import pandas as pd
import sklearn
```

```python
loaded = pd.read_csv("datasets/SongApp.csv", header=-1,)
```

```python
song_app = loaded
```

```python
new_column = {1:"years", 0:"index"}
for i in range(2, 93):
    new_column[i] = "descriptor_" + str(i-1)


song_app = song_app.loc[1:]
song_app.rename(columns=new_column, inplace=True)
song_app
```

```python
nombre_par_annee = song_app.loc[:,"index":"years"].groupby("years").agg('count')
# nombre_par_annee.max
# song_app.drop[]
song_app.drop('index', axis=1, inplace=True)
song_app

# moyenne = song_years.agg('avg')
# moyenne
# truc.agg('avg')
# song_years = song_years.agg('count')
# song_years
```

```python
weird_agg = song_app.groupby("years").agg('count')
weird_agg
weird_agg_2 = weird_agg.loc[:,"descriptor_1"]
weird_agg_2.sort_values()

forbiden = []
for index, value in weird_agg.iterrows():
    if(value[0] < 1000):
        forbiden.append(index)
    print index, value[0]
    
forbiden
# ON FAIT UNE LISTE DE TOUTE LES ANNEES A IGNORER
# weird_agg.loc[1922]
# weird_agg.loc[:,"descriptor_1"]
```

```python
song_app_filtered = song_app[~song_app.years.isin(forbiden)]
song_app
```

```python
pd.DataFrame.hist(song_app_filtered, figsize = [15,15])
```

```python
# NUMBER_OF_ENTRY = 100000
# # we remove the indexes and the year column
# X = song_app.iloc[:NUMBER_OF_ENTRY,1:]
# # we isolate the year column
# y = song_app.iloc[:NUMBER_OF_ENTRY,0]
# X.shape, y.shape

# NUMBER_OF_ENTRY = 10000
NUMBER_OF_ENTRY = 400000

# we remove the indexes and the year column
X = song_app_filtered.iloc[:NUMBER_OF_ENTRY,1:]
# we isolate the year column
y = song_app_filtered.iloc[:NUMBER_OF_ENTRY,0]
X.shape, y.shape
```

```python
X
```

```python
from sklearn.preprocessing import scale
Xs = scale(X)
Xs
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
# from sklearn.cross_validation import train_test_split,cross_val_score v0.15
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import scale
```

# TECHNIQUE 1 : RANDOM FOREST CLASSIFIER

## NO SCALING RANDOMFOREST

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
clf = RandomForestClassifier(n_estimators=300)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyper_params =[{'n_estimators':[20,50,100,200,300]}, {'warm_start':[True,False]} ]
searchres = GridSearchCV(RandomForestClassifier(), hyper_params, scoring='accuracy')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
searchres.fit(X_train,y_train)
```

```python
searchres.best_score_
```

```python
searchres.best_params_
```

```python
print("Performances sur le jeu d'apprentissage:")
print("---")
for params, mean_score, scores in searchres.grid_scores_:
    print("%0.3f (+/-%0.03f) for %r" % (mean_score, scores.std() * 2, params))
```

# TECHNIQUE 3 : K NN


## NO SCALING K-NN

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
from sklearn import neighbors, linear_model

clf = neighbors.KNeighborsClassifier(n_neighbors = 5)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
scores = cross_val_score(clf, X_train, y_train, scoring='accuracy', cv = 5 )
scores.mean()  
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyper_params =[{'n_neighbors': range(2,10)}]
search = GridSearchCV(neighbors.KNeighborsClassifier(), hyper_params, scoring='accuracy')
search.fit(X_train, y_train)
```

```python
search.best_score_
```

```python
search.best_params_
```

```python
print("Performances sur le jeu d'apprentissage:")
print("---")
for params, mean_score, scores in search.grid_scores_:
    print("%0.3f (+/-%0.03f) for %r" % (mean_score, scores.std() * 2, params))
```
