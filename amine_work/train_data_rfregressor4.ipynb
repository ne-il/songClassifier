```python
import pip
import numpy as np
import pandas as pd
import sklearn
import tabulate
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn import preprocessing
```

```python
song_app = pd.read_csv("../../SongApp.csv",header=-1,)
```

```python
song_app.head()
```

```python
# clean data
df_song = song_app.dropna()
df_song.__delitem__(0)
del song_app
```

```python
df_song.head()
```

```python
df_song = df_song[df_song[1] <= 2009]
df_song = df_song[df_song[1] >= 1980]
```

```python
df_song.reindex()
df_song.shape
```

```python
data = df_song.sample(150000)
y_data = data.values[0:,0]
```

```python
data.__delitem__(1)
y_data
```

```python
x_data = data
```

```python
x_data.head()
```

```python
x_data.shape
```

```python
y_data.shape
```

```python
normalized_X = preprocessing.normalize(x_data)
```

```python
# split into training and test data
X_train, X_test, y_train, y_test = train_test_split(normalized_X, y_data, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
clf = RandomForestRegressor(n_jobs=5, n_estimators=40)
model = clf.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import median_absolute_error
```

```python
def log_errors(y_test, predicted):
    mae = mean_absolute_error(y_test, predicted)
    mse = mean_squared_error(y_test, predicted, multioutput='raw_values')
    r2 = r2_score(y_test, predicted)
    medae = median_absolute_error(y_test, predicted)
    print('mae ',mae)
    print('mse ',mse)
    print('r2 score ', r2)
    print('median_ae ',medae)
```

```python
log_errors(y_test=y_test,predicted=predicted)
```

```python
predicted
```

```python
predicted_round = [round(x) for x in predicted]
```

```python
score = 0
for i in range(len(y_test)):
    if(y_test[i] in [predicted[i] -1, predicted[i] + 1]):
        score = score + 1
        print('original : ', y_test[i])
        print('predicted : ', predicted_round[i])
print('score %', score/len(y_test))
print('score ',score)
```
