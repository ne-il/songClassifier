```python
import pip
import numpy as np
import pandas as pd
import sklearn
import tabulate
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
```

```python
song_app = pd.read_csv("../../SongApp.csv",header=-1,)
```

```python
# clean data
df_song = song_app.dropna()
df_song.__delitem__(0)
```

```python
df_song.head()
```

```python
df_song = df_song[df_song[1] <= 2009]
df_song = df_song[df_song[1] >= 1980]
```

```python
labels_column = df_song.iloc[0:,0]
labels = df_song.values[0:,0]
labels
```

```python
import operator
# count by label
from collections import Counter
c = Counter(labels)
d = dict(c)
sorted_labels_size = sorted(d.items(), key=operator.itemgetter(1), reverse=True)
sorted_labels_size[0:60]
```

```python
df_song.__delitem__(1)
```

```python
df_song['labels'] = labels_column
```

```python
num_columns = df_song.columns.shape[0]
columns = list(range(num_columns - 1))
columns_name = list(map(str,columns)) + ['labels']
df_song.columns = columns_name
df_song.head()
```

```python
x_data = df_song
x_data.__delitem__('labels')
```

```python
x_data.head()
```

```python
x_data.shape
```

```python
# split into training and test data
X_train, X_test, y_train, y_test = train_test_split(x_data, labels, test_size=0.1, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
clf = RandomForestRegressor(n_jobs=5, n_estimators=50)
model = clf.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import median_absolute_error
```

```python
def log_errors(y_test, predicted):
    mae = mean_absolute_error(y_test, predicted)
    mse = mean_squared_error(y_test, predicted, multioutput='raw_values')
    r2 = r2_score(y_test, predicted)
    medae = median_absolute_error(y_test, predicted)
    print('mae ',mae)
    print('mse ',mse)
    print('r2 score ', r2)
    print('median_ae ',medae)
```

```python
log_errors(y_test=y_test,predicted=predicted)
```

```python
predicted_round = [round(x) for x in predicted]
```

```python
score = 0
for i in range(len(y_test)):
    if(y_test[i] in [predicted[i] -1, predicted[i] + 1]):
        score = score + 1
        print('original : ', y_test[i])
        print('predicted : ', predicted_round[i])
print('score ', score/len(y_test))
```
