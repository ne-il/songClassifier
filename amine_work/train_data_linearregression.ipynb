```python
# clean data
song_app = song_app.dropna()
song_app.__delitem__(0)
```

```python
labels_column = song_app.iloc[0:,0]
labels = song_app.values[0:,0]
labels
```

```python
song_app.__delitem__(1)
song_app.head()
```

```python
song_app['labels'] = labels_column
```

```python
num_columns = song_app.columns.shape[0]
columns = list(range(num_columns - 1))
columns_name = list(map(str,columns)) + ['labels']
song_app.columns = columns_name
song_app.head()
```

```python
x_data = song_app
x_data.__delitem__('labels')
```

```python
x_data.head()
```

```python
# split into training and test data
X_train, X_test, y_train, y_test = train_test_split(x_data, labels, test_size=0.1, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#least squares method linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print('Estimated intercept coefficient :', lr.intercept_)
```

```python
print('Number of coefficients', len(lr.coef_))
```

```python
# correlation between features and labels
# here we can also remove some useless? features
model = pd.DataFrame(list(zip(X_train.columns, lr.coef_)), columns = ['features', 'estimated coefficients'])
model.sort_values(by='estimated coefficients',ascending=False)
model
```
