```python
import pip
import numpy as np
import pandas as pd
import sklearn
import tabulate
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import median_absolute_error
```

```python
song_app = pd.read_csv("../../SongApp.csv",header=-1,)
```

```python
# clean data
song_app = song_app.dropna()
song_app.__delitem__(0)
```

```python
song_app.head()
```

```python
labels_column = song_app.iloc[0:,0]
labels = song_app.values[0:,0]
labels
```

```python
song_app.__delitem__(1)
song_app.head()
```

```python
song_app['labels'] = labels_column
```

```python
num_columns = song_app.columns.shape[0]
columns = list(range(num_columns - 1))
columns_name = list(map(str,columns)) + ['labels']
song_app.columns = columns_name
song_app.head()
```

```python
x_data = song_app.sample(200000)
y_data = x_data.values[0:,90]
```

```python
y_data
```

```python
x_data.__delitem__('labels')
```

```python
x_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_X = scaler.fit_transform(x_data)
```

```python
scaled_X.shape
```

```python
y_data.shape
```

```python
# split into training and test data
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y_data, test_size=0.1, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#least squares method linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# mean squared error
predict_x_test = lr.predict(X_test)
mse = np.mean((y_test - predict_x_test) ** 2)
print('mse :', mse)
```

```python
preditected_values = [ round(x) for x in predict_x_test]
preditected_values
```

```python
for i in range(len(preditected_values)):
    print('predicted : ',preditected_values[i])
    print('original : ',y_test[i])
```

```python
def log_errors(y_test, predicted):
    mae = mean_absolute_error(y_test, predicted)
    mse = mean_squared_error(y_test, predicted, multioutput='raw_values')
    r2 = r2_score(y_test, predicted)
    medae = median_absolute_error(y_test, predicted)
    print('mae ',mae)
    print('mse ',mse)
    print('r2 score ', r2)
    print('median_ae ',medae)
```

```python
log_errors(y_test=y_test,predicted=preditected_values)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, preditected_values))
```
